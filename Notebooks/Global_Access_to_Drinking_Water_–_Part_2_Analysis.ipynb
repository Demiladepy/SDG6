## Question 4: What is the Annual Rates of Change (ARC) for the national, rural, and urban areas per country?

Annual Rates of Change (ARC) measure how quickly access to basic water services is improving or declining over time. We calculated ARC for three population groups:

- **ARC_n**: National population
- **ARC_r**: Rural population
- **ARC_u**: Urban population


### Key Observations:
- **Rural ARC values** are often higher than urban and national, indicating faster progress in underserved areas.
- **Urban ARC values** tend to be lower, possibly due to saturation in access levels.
- **Negative ARC values** signal a decline in access, which is a critical policy concern.
- **Zero ARC values** often correspond to countries that already have full access (100%).
